```python
import pandas as pd

df = pd.read_csv('syntetic_data.csv')
df
```

```python
answer = df['generated_questions'][0].replace("'", "").strip('][').split(', ')[0]
answer
```

```python
df['original_question'][0]
```

```python
import re
re.sub(r'Вопрос.*Ответ.:', 'Вопрос: ' +answer + ' Ответ: ', df['original_question'][0])
```

```python
question_answers = []
for i in range(len(df)):
    for question in df['generated_questions'][i].replace("'", "").strip('][').split(', '):
        new_sample = re.sub(r'Вопрос.*Ответ.:', 'Вопрос: ' +question + ' Ответ: ', df['original_question'][i])
        question_answers.append(new_sample)
```

```python
len(question_answers)
```

```python
set(question_answers)
```

```python
q_a_syntetic = pd.DataFrame(question_answers,columns=['question_answer'])
q_a_syntetic.drop_duplicates(inplace=True)
q_a_syntetic.to_csv('./data/question_answers_syntetic.csv', index=False)
```
